```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load combined data
df = pd.read_feather("C:\\Users\\mugesh\\Downloads\\Project - Joel\\crypto-forecast-app\\data\\processed\\combined_dataset.feather")

# Convert date if needed
df['date'] = pd.to_datetime(df['date'])

# Overview
df['coin'].value_counts()

# Plot trend of one coin
btc = df[df['coin'] == 'coin_bitcoin']
plt.figure(figsize=(12, 5))
plt.plot(btc['date'], btc['close'], label='Bitcoin Close Price')
plt.title('Bitcoin Price Over Time')
plt.xlabel('Date')
plt.ylabel('Price')
plt.legend()
plt.grid(True)
plt.show()

# Correlation matrix (wide format)
pivoted = df.pivot_table(index='date', columns='coin', values='close')
corr = pivoted.corr()
plt.figure(figsize=(10, 8))
sns.heatmap(corr, annot=False, cmap='coolwarm')
plt.title("Price Correlation Between Coins")
plt.show()
```
