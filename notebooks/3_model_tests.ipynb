```python
import pandas as pd
from prophet import Prophet
import matplotlib.pyplot as plt

# Load Bitcoin data
df = pd.read_feather("C:\\Users\\mugesh\\Downloads\\Project - Joel\\crypto-forecast-app\\data\\processed\\combined_dataset.feather")
btc = df[df['coin'] == 'coin_bitcoin'][['date', 'close']].rename(columns={"date": "ds", "close": "y"})

# Train/test split
btc = btc.dropna()
model = Prophet()
model.fit(btc)

# Forecast
future = model.make_future_dataframe(periods=365)
forecast = model.predict(future)

# Plot forecast
fig1 = model.plot(forecast)
plt.title("Bitcoin Forecast (Prophet)")
plt.show()

# Optional: model.plot_components(forecast)
```
